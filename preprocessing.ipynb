```python
import os
import numpy as np
from scipy.misc import imread, imresize, imsave
from skimage.color import rgb2luv
```

```python
from matplotlib import pyplot as plt
from keras.utils import to_categorical
```

```python
for root,dirs,files in os.walk('Opencountry'):
    for file in files:
        print(str(file))
        img = imresize(imread("Opencountry/"+file, mode='RGB'), (224, 224)).astype(np.float32)
        imsave("Resized/"+file,img)
```

```python
print(len(files))
```

```python
input_images = []
target_values = []
MEAN_VALUES = np.array([104, 117, 123]).reshape((1,1,3))
print(MEAN_VALUES)
for root,dirs,files in os.walk('Resized'):
    for file in files:
        rgb = imread("Resized/"+file, mode='RGB')
        luv = rgb2luv(rgb)
        grayscale = luv[:,:,0]
        target = luv[:,:,1:]
        
        grayscale = np.array(grayscale)
        input_image = np.repeat(grayscale[:, :, np.newaxis],3,axis=2)
        input_image = input_image - MEAN_VALUES
        #print(input_image[:,:,0])
        #print(input_image[:,:,1])
        #print(input_image[:,:,2])
        input_images.append(input_image)
        target_values.append(target)
```

```python
np.save('input_images.npy',input_images)
```

```python
bins = np.linspace(-100,100,num=50)
```

```python
bins
```

```python
u_channel_targets = []
v_channel_targets = []
for target in target_values:
    target = np.array(target)
    quantized_target = np.digitize(target,bins)
    #print(quantized_target)
    #print(np.unique(quantized_target))
    
    u_channel = quantized_target[:,:,0]
    v_channel = quantized_target[:,:,1]
    #print(v_channel.shape)
    
    u_categorical = to_categorical(u_channel, num_classes = 50)
    v_categorical = to_categorical(v_channel, num_classes=50)
    #print(u_categorical.shape)
    #print(u_categorical)
    u_channel_targets.append(u_categorical)
    v_channel_targets.append(v_categorical)
```

```python
np.save('quantized_targets.npy',quantized_targets)
```
